```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from vnstock import Quote
from utilsforecast.plotting import plot_series
from utilsforecast.evaluation import evaluate
from utilsforecast.losses import *

import warnings
warnings.filterwarnings("ignore")
```

## Initial setup

```python
# --- Bước 1: Lấy dữ liệu VNINDEX (Sử dụng code của bạn) ---
print("Đang lấy dữ liệu lịch sử 15 năm của VNINDEX...")
end_date = '2025-10-10' # Điều chỉnh lại ngày cuối cho phù hợp với ngày hiện tại
start_date_object = pd.to_datetime(end_date) - pd.DateOffset(years=15)
start_date_string = start_date_object.strftime('%Y-%m-%d')

quote = Quote(symbol='VNINDEX', source='VCI')
# Lấy dữ liệu và ngay lập tức đặt 'time' làm index
df = quote.history(start=start_date_string, end=end_date)
df.set_index('time', inplace=True)
df.index = pd.to_datetime(df.index)

# Xử lý dữ liệu bị thiếu nếu có
df.dropna(inplace=True)
print("Lấy và xử lý dữ liệu gốc thành công!")
print("Dữ liệu gốc (df):")
print(df.tail())

```

```python

# --- Bước 2: Chuyển đổi df sang định dạng cho statsforecast ---
print("\nĐang chuẩn bị DataFrame cho việc dự báo...")
df_forecast = pd.DataFrame()
df_forecast['ds'] = df.index  # Lấy index (đã là ngày tháng) làm cột 'ds'
df_forecast['y'] = df['close'].values # Lấy giá đóng cửa làm cột 'y'
df_forecast['unique_id'] = "VNINDEX"

df_forecast
```

```python
plot_series(df=df_forecast, ids=["VNINDEX"], palette="viridis")
```

```python
plot_series(df=df_forecast, ids=["VNINDEX"], max_insample_length=56, palette="viridis")
```

## Baseline models

```python
from statsforecast import StatsForecast
from statsforecast.models import Naive, HistoricAverage, WindowAverage, SeasonalNaive
```

```python
horizon = 252 # Dự báo cho 21 ngày tới
models = [
    Naive(),
    HistoricAverage(),
    WindowAverage(window_size=21),
    SeasonalNaive(season_length=25) # Chu kỳ theo tuần
]

sf = StatsForecast(models=models, freq="D") # Tần suất hàng ngày
sf.fit(df=df_forecast)
preds = sf.predict(h=horizon)

print(preds.head())
print(preds.shape)
```

```python
recent_df = df_forecast.tail(252*2) # Lấy 90 ngày gần nhất để vẽ

plt.rcParams.update({'font.size': 12})
plt.figure(figsize=(18, 9))
plt.plot(recent_df['ds'], recent_df['y'], label='Giá thực tế (Observed)', color='blue', linewidth=2.5)

for model_name in preds.columns:
    if model_name not in ['ds', 'unique_id']:
        plt.plot(preds['ds'], preds[model_name], label=f'Dự báo {model_name}', linestyle='--')

last_observed_date = recent_df['ds'].max()
plt.axvline(x=last_observed_date, color='red', linestyle=':', linewidth=1.5, label='Bắt đầu dự báo')

plt.title(f'Dự báo VNINDEX cho {horizon} ngày tới', fontsize=18)
plt.xlabel('Thời gian', fontsize=14)
plt.ylabel('Giá đóng cửa', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
full_df = pd.concat([df_forecast, preds.drop(columns=['unique_id'], errors='ignore')])

# Đặt 'ds' làm index và chuyển sang DatetimeIndex
full_df = full_df.set_index('ds').sort_index()
full_df.index = pd.to_datetime(full_df.index)

# Tùy chỉnh màu sắc cho dễ nhìn
colors = {
    'y': 'blue',
    'Naive': 'darkgreen',
    'HistoricAverage': 'purple',
    'WindowAverage': 'brown',
    'SeasonalNaive': 'cyan'
}

plt.rcParams.update({'font.size': 12})
plt.figure(figsize=(16, 8)) # Tăng kích thước hình
plt.plot(df_forecast['ds'], df_forecast['y'], label='Observed (y)', color=colors['y'], linewidth=2)

for model_name in ['Naive', 'HistoricAverage', 'WindowAverage', 'SeasonalNaive']:
    if model_name in preds.columns:
        # Lọc ra chỉ những giá trị dự báo mới cho model_name
        forecast_values = full_df[model_name].dropna()
        # Chỉ vẽ phần dự báo, không vẽ lại dữ liệu lịch sử
        plt.plot(forecast_values.index, forecast_values, label=f'{model_name} Forecast', color=colors[model_name], linestyle='--', linewidth=1.5)

last_observed_date = df_forecast['ds'].max()
plt.axvline(x=last_observed_date, color='red', linestyle=':', linewidth=1, label='Forecast Start')
plt.title('VNINDEX Price Forecast', fontsize=16) # Tăng kích thước tiêu đề
plt.xlabel('Time', fontsize=14)
plt.ylabel('Price [y]', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10) # Đặt legend ra ngoài

plt.grid(True, linestyle='--', alpha=0.7) # Thêm lưới

plt.xticks(rotation=45, ha='right')

min_val = min(df_forecast['y'].min(), preds[['Naive', 'WindowAverage', 'SeasonalNaive']].min().min()) # Bỏ qua HistoricAverage nếu nó quá thấp
max_val = max(df_forecast['y'].max(), preds[['Naive', 'WindowAverage', 'SeasonalNaive']].max().max()) # Bỏ qua HistoricAverage nếu nó quá thấp

# Thêm một khoảng đệm nhỏ
y_padding = (max_val - min_val) * 0.1
plt.ylim(min_val - y_padding, max_val + y_padding)


plt.tight_layout(rect=[0, 0, 0.85, 1]) # Điều chỉnh layout để legend không bị cắt
plt.show()
```

### Evaluate baseline models

```python
from utilsforecast.losses import mae, mse # Import các hàm tính lỗi
from utilsforecast.evaluation import evaluate # Import hàm evaluate mới

test = df.iloc[-252:].reset_index(drop=True)  # Lấy 21 ngày cuối làm tập test
train = df.iloc[:-252].reset_index(drop=True) # Phần còn lại làm tập train

# Convert 'ds' column to numeric (timestamp) for cross_validation
df_forecast['ds_num'] = df_forecast['ds'].astype('int64') // 10**9
\
crossval_df = sf.cross_validation(
    df=df_forecast.drop('ds', axis=1).rename(columns={'ds_num': 'ds'}),
    h=horizon,
    step_size=horizon, # Bước nhảy bằng số ngày dự báo
    n_windows=5      # Thực hiện 5 lần kiểm tra trên các khoảng thời gian khác nhau
)

print("\nKết quả Cross-Validation:")
print(crossval_df.head())

evaluation_df = pd.merge(
    crossval_df,
    df_forecast.rename(columns={'y': 'y_true'}), # Đổi tên cột y thành y_true
    on=['ds', 'unique_id'],
    how='left'
)

evaluation_df['ds'] = pd.to_datetime(evaluation_df['ds'], unit='s')

evaluation_summary = evaluate(
    df=evaluation_df,
    metrics=[mae, mse]
)

print(evaluation_summary)
```

```python
sf.fit(df=df_forecast.iloc[:-21])

preds = sf.predict(h=horizon)

# Create a test DataFrame with 'ds' and 'unique_id'
test = df_forecast.tail(21)

eval_df = pd.merge(test, preds, on=['ds', 'unique_id'], how='left')
```

```python
evaluation = evaluate(
    eval_df,
    metrics=[mae],
)
evaluation
```

## Forecasting with exogenous features

```python
df = pd.read_csv("data/daily_sales_french_bakery.csv", parse_dates=["ds"])
df = df.groupby('unique_id').filter(lambda x: len(x) >= 28)
df.head()
```

```python
baguette_plot_df = df[df["unique_id"] == "BAGUETTE"]
croissant_plot_df = df[df["unique_id"] == "CROISSANT"]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(16,8))

ax1.plot(baguette_plot_df["ds"], baguette_plot_df["y"])
ax1.set_xlabel("Date")
ax1.set_ylabel("Baguette sales volume")

ax2.plot(baguette_plot_df["ds"], baguette_plot_df["unit_price"])
ax2.set_xlabel("Date")
ax2.set_ylabel("Unit price of baguette")

ax3.plot(croissant_plot_df["ds"], croissant_plot_df["y"])
ax3.set_xlabel("Date")
ax3.set_ylabel("Croissant sales volume")

ax4.plot(croissant_plot_df["ds"], croissant_plot_df["unit_price"])
ax4.set_xlabel("Date")
ax4.set_ylabel("Unit price of croissant")

fig.autofmt_xdate()
plt.tight_layout()
```

```python
unique_ids = ["BAGUETTE", "CROISSANT"]
small_df = df[df["unique_id"].isin(unique_ids)]
test = small_df.groupby("unique_id").tail(7)
train = small_df.drop(test.index).reset_index(drop=True)
```

```python
futr_exog_df = test.drop(["y"], axis=1)
futr_exog_df.head()
```

```python
test_df = test.merge(arima_exog_preds, on=["unique_id", "ds"], how="left")\
              .merge(arima_preds, on=["unique_id", "ds"], how="left")
test_df
```

## Evaluation metrics

```python
models = [
    AutoARIMA(season_length=7, alias="SARIMA_exog"),
    SeasonalNaive(season_length=7)
]

sf = StatsForecast(models=models, freq="D")
final_cv_df = sf.cross_validation(
    h=horizon,
    df=small_df,
    n_windows=8,
    step_size=7,
    refit=True,
    level=[80],
)
```

```python
final_cv_df.head()
```

```python
temp_test = small_df.groupby("unique_id").tail(7*8)
temp_train = small_df.drop(temp_test.index).reset_index(drop=True)
```

```python
models = ["SARIMA_exog", "SeasonalNaive"]
metrics = [
    mae,
    mse, 
    rmse, 
    mape, 
    smape,
    partial(mase, seasonality=7),
    scaled_crps
]

final_eval = evaluate(
    final_cv_df.drop(["ds", "cutoff"], axis=1),
    metrics=metrics,
    models=models,
    train_df=temp_train,
    level=[80]
)
final_eval = final_eval.drop(['unique_id'], axis=1).groupby('metric').mean().reset_index()
final_eval
```

```python
fig, axes = plt.subplots(3, 3, figsize=(18, 12))

axes_flat = axes.flatten()

models = ['SARIMA_exog', 'SeasonalNaive']
x_pos = [0, 1]
colors = ['blue', 'red']

for i, row in final_eval.iterrows():
    ax = axes_flat[i]
    
    model_values = [row['SARIMA_exog'], row['SeasonalNaive']]
    
    bars = ax.bar(x_pos, model_values, color=colors, alpha=0.8, edgecolor='black', linewidth=1)
    
    for j, (bar, value) in enumerate(zip(bars, model_values)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                f'{value:.3f}', ha='center', va='bottom', fontweight='bold', fontsize=10)
    
    ax.set_title(row['metric'].upper(), fontweight='bold', fontsize=12)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(models, ha='center')
    ax.set_ylabel('Value')
    max_value = max(model_values)
    ax.set_ylim(0, max_value * 1.1)

fig.delaxes(axes_flat[7])

axes_flat[8].set_visible(False)

plt.tight_layout()
plt.show()
```
